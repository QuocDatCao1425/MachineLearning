```python
import warnings
warnings.filterwarnings(action = 'ignore')
from PIL import Image
import matplotlib.pyplot as plt

from transformers import pipeline
from transformers.utils import logging
logging.set_verbosity(60)

from IPython.display import Image

import os



```

```python
from google.colab import drive
drive.mount('/content/drive')
print(os.listdir("/content/drive/My Drive/"))  # Check for "Notebook"
print(os.listdir("/content/drive/My Drive/Notebook/"))

```

```python


logging.set_verbosity(60)  # Suppress warnings

# Mount Google Drive to access images
drive.mount('/content/drive')

# Define image paths inside "Notebook" folder
image_paths = [
    "/content/drive/My Drive/Notebook/a1.jpeg",
    "/content/drive/My Drive/Notebook/a2.png",
    "/content/drive/My Drive/Notebook/a3.png"
]

# Load the image captioning pipeline
captioner = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")

# Function to generate caption
def generate_caption(image_path):
    image = Image.open(image_path).convert("RGB")  # Ensure RGB mode
    caption = captioner(image)[0]['generated_text']
    return image, caption

# Display images with captions
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, image_path in enumerate(image_paths):
    if os.path.exists(image_path):  # Check if file exists
        image, caption = generate_caption(image_path)
        axes[i].imshow(image)
        axes[i].set_title(caption, fontsize=12)
        axes[i].axis("off")
    else:
        axes[i].set_title("File Not Found", fontsize=12)
        axes[i].axis("off")

plt.show()
```
